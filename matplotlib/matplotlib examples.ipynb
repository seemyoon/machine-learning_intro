# plot and scatter diagrams

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
values = [0, 1, 2, 3, 4, 5, 6, 7]
```

```python
squares = [pow(value, 2) for value in values]
plt.plot(values, squares)
```

```python
values = list(range(100))
cubes = [pow(value, 3) for value in values]
plt.scatter(values, cubes)
plt.show()
```

```python
values = list(range(30))
color_p = [(pow(value, 3)) for value in values]
plt.scatter(values, color_p, c=color_p, cmap=plt.cm.Reds, edgecolors='none', s=20)
plt.show()
```

```python
values2 = list(range(40))
squares2 = [pow(value, 2) for value in values2]
cubes2 = [pow(value, 3) for value in values2]
fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True)
ax1.scatter(values2, squares2)
ax1.set_title('Square numbers')
ax2.scatter(values2, cubes2)
ax2.set_title('Cube numbers')
plt.show()
```

# Using DataFrames for Creating Diagrams

```python
data = {
    'year': [2001, 2002, 2003, 2004, 2005, 2006],
    'sales': [100, 150, 130, 180, 200, 210]
}
df = pd.DataFrame(data)
df.plot(x='year', y='sales', kind='line')
plt.title('yearly sales')
plt.show()
```

```python
data = {
    'age': [10, 40, 45, 44, 50, 70, 80, 100]
}
df = pd.DataFrame(data)
df.plot(y='age', kind='hist')
plt.show()
```

```python
temp_data = {
    'day': ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'],
    'temperature': [22, 24, 19, 21, 23, 25, 26]
}
df = pd.DataFrame(temp_data)
df.plot(x='day', kind='line', y='temperature', marker='o')
plt.title('Weekly Temperature')
plt.show()
```

```python
data = {
    'year': [2001, 2002, 2003, 2004, 2005, 2006],
    'sales': [100, 150, 130, 180, 200, 210]
}
df = pd.DataFrame(data)
df.plot(y='sales', labels=df['year'], kind='pie', autopct='%1.1f%%')
plt.title('yearly sales')
plt.show()
```

# Boxplots

```python
data = {
    'Category': ['FIRST', 'FIRST', 'FIRST', 'SECOND', 'SECOND', 'SECOND', 'THIRD', 'THIRD', 'THIRD'],
    'Value': [100, 150, 130, 180, 200, 210, 300, 310, 400]
}
df = pd.DataFrame(data)
df
```

```python
df.boxplot(by="Category", column='Value')
plt.title('Values distribution by category')
plt.show()
```

# outliers

```python
# visually
data = {
    'Company': ['Apple', 'Apple', 'Apple', 'Apple', 'Apple', 'Apple', 'Microsoft', 'Microsoft', 'Microsoft',
                'Microsoft', 'Microsoft', 'Microsoft', 'Tesla', 'Tesla', 'Tesla', 'Tesla', 'Tesla', 'Tesla'],
    'Salary': [5000, 5200, 5500, 4800, 5100, 5200, 20000, 22000, 25000, 30000, 35000, 40000, 6000, 6200, 6300, 5900,
               6100, 5900]
}

df = pd.DataFrame(data)
df.plot(x='Company', y='Salary', kind='line', marker='o')
plt.title('Weekly Temperature')
plt.show()
```

```python
df.boxplot(by='Company', column='Salary')
plt.title('Salary distribution by company')
plt.show()
```

```python
df_first = df[df['Company'] == 'Apple']
df_first.boxplot(by='Company', column='Salary')
plt.title('Values distribution for FIRST category')
plt.show()
```

```python
median = df['Salary'].median()

Q1 = df['Salary'].quantile(0.25)
Q3 = df['Salary'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print("Lower bound:", lower_bound)
print("Upper bound:", upper_bound)
print("Median:", median)
```

# Heatmap

```python
data_temp2 = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May'],
    'Week 1 temperature': [-1, 4, 10, 14, 17],
    'Week 2 temperature': [0, 6, 11, 16, 18],
    'Week 3 temperature': [-2, 5, 14, 17, 20],
    'Week 4 temperature': [-1, 7, 15, 17, 21],
}
df_temp = pd.DataFrame(data_temp2)
df_temp.head()
```

```python
df_temp = df_temp.set_index('Month')
df_temp.head()
```

```python
sns.heatmap(df_temp, annot=True, cmap='coolwarm')
plt.title('Weekly temperature data')
plt.show()
```
