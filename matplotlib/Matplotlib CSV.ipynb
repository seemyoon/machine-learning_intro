```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('jersey-rainfall-1984-to-2021.csv', index_col=0)
df.head(10)
```

```python
df.plot(y='YearTotalmm')
plt.show()
```

```python
df.plot(y='Jan')
plt.show()
```

```python
df.plot(y=['Jan', 'Feb'], kind='area')
plt.show()
```

```python
df.drop(columns=['YearTotalmm']).plot()
plt.title('Title without YearTotalmm')
plt.show()
```

```python
df.T.head()
```

```python
df.T.plot(y=1901)
```

```python
df.T.plot(y=[1910,1912])
```

```python
df.drop(columns=['YearTotalmm']).T.plot(y=[1910, 1912])
plt.show()
```

```python
df.drop(columns=['YearTotalmm']).T.plot(y=range(1910, 1913))
plt.show()
```

```python
df.drop(columns=['YearTotalmm']).T.plot(y=range(2010, 2014), kind='area')
plt.show()
```
